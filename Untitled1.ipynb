```python
import numpy as np
import pandas as pd
from math import sqrt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.gaussian_process import GaussianProcessRegressor
from sklearn.metrics import mean_absolute_error
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv(r"market_data.csv")
```

```python
data
```

```python
X=data.iloc[:,2:11]
Y1=data.iloc[:,11]
Y2=data.iloc[:,12]
X
```

```python
Y1
```

```python
Y2
```

```python
X1_train, X1_test, Y1_train, Y1_test = train_test_split(X, Y1, test_size=0.4, random_state=1) 
X2_train, X2_test, Y2_train, Y2_test = train_test_split(X, Y2, test_size=0.4, random_state=1) 
```

```python
#Fitting MLR to X1_train and Y1_train
reg1 = LinearRegression()
reg1.fit(X1_train,Y1_train)
Y1_pred = reg1.predict(X1_test)
Y1_pred
```

```python
print(sqrt(mean_squared_error(Y1_test, Y1_pred)))
print(mean_absolute_error(Y1_test, Y1_pred))
plt.scatter(Y1_test, Y1_pred)
plt.show()
```

```python
#Fitting MLR to X2_train and Y2_train
reg2 = LinearRegression()
reg2.fit(X2_train,Y2_train)
Y2_pred = reg2.predict(X2_test)
Y2_pred
```

```python
print(sqrt(mean_squared_error(Y2_test, Y2_pred)))
print(mean_absolute_error(Y2_test, Y2_pred))
plt.scatter(Y2_test, Y2_pred)
plt.show()
```

```python
#Fitting GPR to X1_train and Y1_train
gpr1 = GaussianProcessRegressor(kernel=None, random_state=0)
gpr1.fit(X1_train,Y1_train)
gprY1_pred = gpr1.predict(X1_test)
gprY1_pred
```

```python
print(sqrt(mean_squared_error(Y1_test, gprY1_pred)))
print(mean_absolute_error(Y1_test, gprY1_pred))
plt.scatter(Y1_test, gprY1_pred)
plt.show()
```

```python
#Fitting GPR to X2_train and Y2_train
gpr2 = GaussianProcessRegressor(kernel=None, random_state=0)
gpr2.fit(X2_train,Y2_train)
gprY2_pred = gpr2.predict(X2_test)
gprY2_pred
```

```python
print(sqrt(mean_squared_error(Y2_test, gprY2_pred)))
print(mean_absolute_error(Y2_test, gprY2_pred))
plt.scatter(Y2_test, gprY2_pred)
plt.show()
```
